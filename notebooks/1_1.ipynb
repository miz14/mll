```python

data = data.head(200000)
data.drop(['id','pickup_datetime', 'dropoff_datetime', 'vendor_id', 'passenger_count',  ], axis=1, inplace=True)
data.drop_duplicates(inplace=True)
data.dropna(inplace=True)
```

```python
data.to_csv("../data/trip_duration_task_m.csv")
```

```python
data.info()
```

Предобработка данных

1) Убрал строки с null-value в столбце "vendor_id"
2) Заполнил строки с null-value в столбце "pickup_latitude"

```python
y = data['trip_duration']
X = data.drop(['trip_duration'], axis=1)

```

```python
y
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
lr = LinearRegression().fit(X_train, y_train)
```

```python
lr.predict(X_test)
list_column_weight = list(zip(X.columns,lr.coef_))
list_column_weight
```

```python
y_pred = lr.predict(X_test)
y_pred.shape
```

```python
print(f'MAE: {mean_absolute_error(y_test, y_pred)}')
print(f'MSE: {mean_squared_error(y_test, y_pred)}')
print(f'RMSE: {sqrt(mean_squared_error(y_test, y_pred))}')
print(f'MAPE: {sqrt(mean_absolute_percentage_error(y_test, y_pred))}')
print(f'R^2: {r2_score(y_test, y_pred)}')
```

```python
len(lr.coef_)
lr.coef_
```

```python
ridge = Ridge(alpha=0.9).fit(X_train, y_train)
y_pred = ridge.predict(X_test)
print(f'MAE: {mean_absolute_error(y_test, y_pred)}')
print(f'MSE: {mean_squared_error(y_test, y_pred)}')
print(f'RMSE: {sqrt(mean_squared_error(y_test, y_pred))}')
print(f'MAPE: {sqrt(mean_absolute_percentage_error(y_test, y_pred))}')
print(f'R^2: {r2_score(y_test, y_pred)}')
ridge.coef_
```

```python
parameters = {'alpha': np.arange(0, 1, 0.1)}
ridge_optimal = GridSearchCV(Ridge(), parameters).fit(X_train, y_train)
ridge_optimal.best_params_
```

```python
lasso = Lasso(alpha=0.5).fit(X_train, y_train)
y_pred = lasso.predict(X_test)
print(f'MAE: {mean_absolute_error(y_test, y_pred)}')
print(f'MSE: {mean_squared_error(y_test, y_pred)}')
print(f'RMSE: {sqrt(mean_squared_error(y_test, y_pred))}')
print(f'MAPE: {sqrt(mean_absolute_percentage_error(y_test, y_pred))}')
print(f'R^2: {r2_score(y_test, y_pred)}')
lasso.coef_
```

```python
polynomialFeatures = PolynomialFeatures(2)
X = data[['dropoff_latitude','dropoff_longitude', 'pickup_latitude', 'pickup_longitude']].values.reshape(-1, 4)
y = data['trip_duration'].values.reshape(-1, 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
lr = LinearRegression().fit(X_train, y_train)
lr.coef_
```

```python
X_pf = polynomialFeatures.fit_transform(X_train)
X_pf_test = polynomialFeatures.fit_transform(X_test)
```

```python
lr2 = LinearRegression().fit(X_pf, y_train)
y_pred_pf = lr2.predict(X_pf_test)
print(f'MAE: {mean_absolute_error(y_test, y_pred_pf)}')
print(f'MSE: {mean_squared_error(y_test, y_pred_pf)}')
print(f'RMSE: {sqrt(mean_squared_error(y_test, y_pred_pf))}')
print(f'MAPE: {sqrt(mean_absolute_percentage_error(y_test, y_pred_pf))}')
print(f'R^2: {r2_score(y_test, y_pred_pf)}')
```

```python
ridge_pf = Ridge(alpha=0.9).fit(X_pf, y_train)
y_pred_pf = ridge_pf.predict(X_pf_test)
print(f'MAE: {mean_absolute_error(y_test, y_pred_pf)}')
print(f'MSE: {mean_squared_error(y_test, y_pred_pf)}')
print(f'RMSE: {sqrt(mean_squared_error(y_test, y_pred_pf))}')
print(f'MAPE: {sqrt(mean_absolute_percentage_error(y_test, y_pred_pf))}')
print(f'R^2: {r2_score(y_test, y_pred_pf)}')
```

```python
lasso_pf = Lasso(alpha=0.9).fit(X_pf, y_train)
y_pred_pf = lasso_pf.predict(X_pf_test)
print(f'MAE: {mean_absolute_error(y_test, y_pred_pf)}')
print(f'MSE: {mean_squared_error(y_test, y_pred_pf)}')
print(f'RMSE: {sqrt(mean_squared_error(y_test, y_pred_pf))}')
print(f'MAPE: {sqrt(mean_absolute_percentage_error(y_test, y_pred_pf))}')
print(f'R^2: {r2_score(y_test, y_pred_pf)}')
```
